```python
import os

adjlist_dir = '/content/pscan/data/adjlists'

for filename in sorted(os.listdir(adjlist_dir)):
    if filename.endswith('.adjlist'):
        input_path  = f'{adjlist_dir}/{filename}'
        output_path = f'{adjlist_dir}/{filename.replace(".adjlist", ".sim.tsv")}'
        print(f"\n{'='*50}")
        print(f"Running: {filename}")
        print(f"{'='*50}")
        os.system(f'python similarity_main.py --input {input_path} --output {output_path} --verbose')
```

```python
print("Generated similarity files:")
for f in sorted(os.listdir(adjlist_dir)):
    if f.endswith('.sim.tsv'):
        size = os.path.getsize(f'/content/pscan/data/adjlists/{f}')
        print(f"  {f}  →  {size/1024/1024:.2f} MB")
```
